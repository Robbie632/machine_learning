# RFE

```python
filepath = 'final_figures/RFE.png'

RFCV_df = pd.read_csv('output_datasets/pre_model_datasets/RFCV_DF.csv')
```

```python
sns.set(rc={'figure.figsize':(4,4)})
sns.set(font_scale=1.8)
plot = sns.scatterplot(data = RFCV_df, x = 'Number of Features', y = 'F1 Score', s = 70)

plt.title('Features evaluated by F1 score', fontsize = 30)

if save_plots:
    fig = plot.get_figure()
    fig.savefig(filepath)
```

# learning curve

```python


filepath = 'final_figures/LC.png'

rfc_lc = pd.read_csv('output_datasets/pre_model_datasets/rfc_lc.csv')
```

```python
sns.set(rc={'figure.figsize':(50,40)})
sns.set(style="ticks", rc={"lines.linewidth": 7.0})

plot = sns.lineplot(x = 'train_sizes', y = 'k_fold_stratified_scores', hue = 'train_or_test_data', ci = 'sd', data = rfc_lc, markers = True)
plot.set_xlabel("train fold size (number of observations)",fontsize=50)
plot.set_ylabel("F1 score",fontsize=70)
plot.set_title('learning curve', fontsize = 80)
plot.set_xlabel("train fold size (number of observations)",fontsize=70)
plot.tick_params(labelsize=40)

plt.legend(fontsize = 50, loc = 4)

if save_plots == True:
    fig = plot.get_figure()
    fig.savefig(filepath)
```
