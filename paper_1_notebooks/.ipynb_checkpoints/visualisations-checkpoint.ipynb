# final visualisations for paper

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# tsne bedrock

```python
save_plots = True
```

```python
filepath = 'final_figures/bedrock_tsne.eps'
tsne_data_with_colour = pd.read_csv('output_datasets/pre_model_datasets/tsne_data_with_color_bedrock.csv')
```

```python

sns.set(rc={'figure.figsize':(2.4,2.4)})


plot_on_x = 'tsne1'
plot_on_y = 'tsne2'
sns.set(font_scale=0.5)
plot = sns.scatterplot(data = tsne_data_with_colour, linewidth = 0.3, edgecolor = 'black', s= 2.8, x = plot_on_x, y = plot_on_y, hue='class',
                palette=sns.color_palette("Paired", tsne_data_with_colour['class'].nunique()))

plt.title('t-SNE Analysis of Bedrock Sample Data', fontsize = 6)
plt.legend(fontsize=2.5, markerscale=0.07)
plt.yticks(fontsize = 5)
plt.xticks(fontsize = 5)
plt.tight_layout()

if save_plots == True:
    fig = plot.get_figure()
    fig.savefig(fname=filepath, dpi = 2000, pad_inches = 0.5)
    
```

# tsne outliers

```python
filepath = 'final_figures/outlier_tsne.eps'

tsne_data_with_colour = pd.read_csv('output_datasets/pre_model_datasets/tsne_outlier_data.csv')
```

```python
sns.set(rc={'figure.figsize':(3.8,3.8)})

plot_on_x = 'tsne1'
plot_on_y = 'tsne2'

sns.set(font_scale=0.8)
plot = sns.scatterplot(data = tsne_data_with_colour, s= 8, x = plot_on_x, y = plot_on_y, hue='inlierLabel', linewidth = 0.3, edgecolor = 'black',
                palette=sns.color_palette("muted", tsne_data_with_colour['inlierLabel'].nunique())
               )
plt.title('t-SNE Analysis of Superficial Sample Data', fontsize = 8)
plt.legend(fontsize=5, markerscale=0.16)
plt.yticks(fontsize = 7)
plt.xticks(fontsize = 7)
plt.tight_layout()


if save_plots == True:
    fig = plot.get_figure()
    fig.savefig(fname=filepath, dpi = 2000, pad_inches = 0.5)
```

# RFE

```python
filepath = 'final_figures/RFE.eps'

RFCV_df = pd.read_csv('output_datasets/pre_model_datasets/RFCV_DF.csv')
```

```python
sns.set(rc={'figure.figsize':(4,4)})
sns.set(font_scale=0.6)

plot = sns.scatterplot(data = RFCV_df, x = 'Number of Features', y = 'F1 Score', s = 30)

plt.title('Features evaluated by F1 score', fontsize = 8)


plt.yticks(fontsize = 7)
plt.xticks(fontsize = 7)
plt.tight_layout()

if save_plots:
    fig = plot.get_figure()
    fig.savefig(filepath)
```

# learning curve

```python


filepath = 'final_figures/LC.png'

rfc_lc = pd.read_csv('output_datasets/pre_model_datasets/rfc_lc.csv')
```

```python
sns.set(rc={'figure.figsize':(50,40)})
sns.set(style="ticks", rc={"lines.linewidth": 7.0})

plot = sns.lineplot(x = 'train_sizes', y = 'k_fold_stratified_scores', hue = 'train_or_test_data', ci = 'sd', data = rfc_lc, markers = True)
plot.set_xlabel("train fold size (number of observations)",fontsize=50)
plot.set_ylabel("F1 score",fontsize=70)
plot.set_title('learning curve', fontsize = 80)
plot.set_xlabel("train fold size (number of observations)",fontsize=70)
plot.tick_params(labelsize=40)

plt.legend(fontsize = 50, loc = 4)

if save_plots == True:
    fig = plot.get_figure()
    fig.savefig(filepath)
```
